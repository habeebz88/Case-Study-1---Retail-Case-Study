```python
def dateconvert(date):
    return dateparser.parse(date)
customer_final["tran_date"] = customer_final["tran_date"].apply(dateconvert)
customer_final["tran_date"]
```

```python
customer_final["DOB"] = pd.to_datetime(customer_final["DOB"],format = "%d-%m-%Y")
customer_final["tran_date"] = pd.to_datetime(customer_final["tran_date"])
customer_final["Age"] = (customer_final["tran_date"] - customer_final["DOB"])
```

```python
customer_final["Age"] = customer_final["Age"] / 365.25
```

```python
customer_final["Age"] = customer_final["Age"].dt.components["days"]
```

```python
age_between = customer_final[(customer_final["Age"] >= 25) & (customer_final["Age"] <= 35)]
age_between.head(5)
```

```python
filter_data = age_between[age_between["prod_cat"].isin(["Electronics","Books"])]
```

```python
total = filter_data.groupby("prod_cat")["total_amt"].sum()
total = round(total,2)
print("the total amount spent for “Electronics” and “Books” are:")
print(total)
```

```python
date_fil = age_between[(age_between["tran_date"] >= "2014-01-01") & (age_between["tran_date"] <= "2014-03-01")]
```

```python
tot = date_fil["total_amt"].sum()
tot = round(tot,2)
print("he total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014: ",tot)
```
